```python
import numpy as np
from pym import func as pym
from pyg.colors import pu as puc

R = 8.314
a = 5.464
b = 30.5
T_r = 7.0 / 8.0
v_r = np.linspace(0.5, 8.0, 1000)
p_r = ((8.0*T_r) / (3.0 * v_r - 1.0)) - (3.0 / np.power(v_r, 2.0))

vdw = pym.curve(v_r, p_r, name=r'Isotherm at reduced temperature ($T_{r}=\frac{T}{T_{c}}=\frac{7}{8}$) [ ]')

plot = vdw.plot(linestyle='-', linecolor=puc.pu_colors['blue'])
crossings = vdw.find(0.6)
vdw_pneg = vdw.copy().crop(x_min=crossings[0], x_max=crossings[1], replace='remove')
plot.fill_between(vdw_pneg.x, 0.6 * np.ones_like(vdw_pneg.y), vdw_pneg.y, fc=puc.pu_colors['lightblue'],
                  name='Negative Pressure Region')
vdw_sheat = vdw.copy().crop(x_min=crossings[1], x_max=crossings[2], replace='remove')
plot.fill_between(vdw_sheat.x, 0.6 * np.ones_like(vdw_sheat.y), vdw_sheat.y, fc=puc.pu_colors['orange'],
                  name='Superheat Region')
plot.add_hline(0.6, crossings[0], crossings[2], ls='--', color=puc.pu_colors['purple'])
plot.add_text((crossings[0] + crossings[2])/2.0, 0.6, 'Coexistance Pressure', ha='center', va='bottom',
              color=puc.pu_colors['purple'])
plot.add_vmeasure(crossings[0] + (crossings[1] - crossings[0])/3.0, 0.6, 0.45, '$p_{neg}$')
plot.lines_on()
plot.markers_off()
plot.ylim(0.2, 0.75)
plot.xlim(0.0, 4.0)
plot.xlabel(r'Reduced molar volume ($v_{r}=\frac{v}{v_{c}}$) [ ]')
plot.ylabel(r'Reduced pressure ($p_{r}=\frac{p}{p_{c}}$) [ ]')
plot.legend()

plot.export('img/vdw')
plot.show('Van Der Waals Isotherm for Water at $25\unit{^{o}C}$', label='van-der-waals')


lyx.todo('plot the van der waals loop')
```
